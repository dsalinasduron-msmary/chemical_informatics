```python
#working in Jupyter Notebook with Ubuntu
import deepchem as dc
from deepchem.models import GCNModel
import pandas as pd
import os
import requests
import numpy
print("imported")

working_dir = os.getcwd()
#the_directory = os.path.join(working_dir, "0000001_0001000")
#os.chdir(the_directory)
print(os.getcwd())

#get dataset from computer
print("WORKING FROM GCN MODEL")

!find -name '*.csv' -print0 | xargs -0 -n1 chmod 755

orig_df = pd.read_csv("736.csv", sep = ",", header = 0)

print(orig_df.shape)
print(orig_df.columns)




```

```python
#top 2 rows are not part of the data; get rid of them
the_df = orig_df[pd.to_numeric(orig_df['PUBCHEM_RESULT_TAG'], errors='coerce').notnull()]
the_df.head(30)
```

```python
#Active compounds marked with 1, inactive marked with 0
the_df["ACTIVE?"] = the_df.PUBCHEM_ACTIVITY_OUTCOME.replace({"Active" : 1, "Inactive" : 0})
the_df.head(100)
#del the_df["PUBCHEM_ACTIVITY_OUTCOME"]
```

```python
the_df.head(100)
```

```python
the_df.rename(columns ={"PUBCHEM_EXT_DATASOURCE_SMILES" : "SMILES"}, inplace = True)
the_df.head(100)
```

```python
#featurize smiles to prep for model training
smiles = the_df["SMILES"]
labels = the_df["ACTIVE?"]
featurizer = dc.feat.MolGraphConvFeaturizer()
X = featurizer.featurize(smiles)
the_df["FEATURIZED"] = X
```

```python
#take featurized dataset and convert to csv
ofile = "736_featurized.csv"
the_df.to_csv(ofile, index = False)
```

```python
!pwd
```

```python
#take subset of data for training
train_df = the_df.sample(frac = 0.7)
train_df.head(100)
```

```python
#take other subset of data for testing
test_df = the_df[~the_df["PUBCHEM_RESULT_TAG"].isin(train_df["PUBCHEM_RESULT_TAG"])]
test_df.head(100)
```

```python
#format data into 2D NumPy array
train_dataset = dc.data.NumpyDataset(X=train_df["FEATURIZED"], y = train_df["ACTIVE?"])
test_dataset = dc.data.NumpyDataset(X=test_df["FEATURIZED"], y = test_df["ACTIVE?"])
```

```python
#check shape
print(train_dataset)
print(test_dataset)
print(the_df.shape)
```

```python
#train model. This should take ~45 seconds per 100,000 datapoints per epoch on this CPU. 
model = GCNModel(mode='classification', n_tasks=1,
                 batch_size=16, learning_rate=0.001)
model.fit(train_dataset, nb_epoch=50)
print('done')
```

```python
#Get ROC AUC to test accuracy of model
metric = dc.metrics.Metric(dc.metrics.roc_auc_score)
print('Training set score:', model.evaluate(train_dataset, [metric]))
print('Test set score:', model.evaluate(test_dataset, [metric]))
```
